```python
import numpy as np

import random

import matplotlib.pyplot as plt
import cv2
```

```python
def maxtanh(x,s=0.25,alpha=0.3):
    #print(x)
    t=np.tanh(x/s)
    #print(t)
    c=np.array((t,x*s))
    #print(c)
    m=c.max(axis=0)
    n=c.min(axis=0)
    #print(m)
    z=np.zeros_like(m)
    return np.max((m,z),axis=0)+np.min((n,z),axis=0)*alpha
    #return np.max((m,z),axis=0)
def xtanh(x,s=10,alpha=0.3):
    t=np.tanh(x)
    r=t+x/s
    c=np.array((r,r*alpha))
    m=c.max(axis=0)
    return m
def dtanh(x):
    t=np.tanh(x)
    t2=t*t
    dt=np.ones_like(t2)-t2
    return dt
```

```python
x=np.arange(-3.0,3.0,0.01)
s=0.6
y=maxtanh(x,s)
plt.figure(figsize=(10,10))
line, = plt.plot(x, y, lw=2)
#plt.plot(x,x)
plt.plot(x,np.tanh(x))
#plt.plot(x,x*s)
#plt.plot(x,np.tanh(x/s))
#plt.plot(x,dtanh(x/s))
plt.plot(x,xtanh(x))
#plt.plot(x,np.ones_like(x)*dtanh(1.15/s))
plt.plot(x,np.zeros_like(x))
plt.show()
```
